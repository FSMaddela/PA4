# PART 1: Use Data Wrangling and Visualization technique to analyze the dataframe

**Import pandas as pd and load the xlsx file.**

```python
import pandas as pd

df = pd.read_excel("board2.xlsx")
```

**Create a new column for the Average**

```python
df['Average'] = df[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1)
df
```

**Create a dataframe 'instru' of track instrumentation, hometown Luzon, and a grade greater than 70 in electronics**

```python
instru = df[(df["Track"] == "Instrumentation") & 
            (df["Hometown"] == "Luzon") & 
            (df["Electronics"] > 70)]

instru = instru[["Name", "GEAS", "Electronics"]]
instru
```

**Create a dataframe 'mindy' of hometown Luzon, gender female, and an average greater than or equal to 55**

```python
mindy = df[(df["Hometown"] == "Mindanao") & 
           (df["Gender"] == "Female") & 
           (df["Average"] >= 55)]


mindy = mindy[["Name", "Track", "Electronics", "Average"]]
mindy
```

# Part 2: Create a visualization to determine whether demographical factors affect the average scores

**Import pandas as pd and load the xlsx file.**

```python
df = pd.read_excel("board2.xlsx")
```

**Create a new column for the Average**

```python
df['Average'] = df[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1)
df
```

**Use seaborn and matplotlib to construct a bar graph of the data**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


plt.figure(figsize=(10, 10))


sns.barplot(x="Track", y="Average", hue="Gender", data=df, errorbar=None)


plt.title("Average Grade by Track, Gender, and Hometown")
plt.xlabel("Track")
plt.ylabel("Average Grade")


plt.xticks(rotation=45)


plt.show()
```
